```python
import matplotlib.pyplot as plt
print(plt.__package__) #set the package name
```

```python
plt.figure(figsize=(6,4))
plt.plot([1,2,3,4,5],[4,5,6,7,8])
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.plot(x,y, marker='o', linestyle='-', color='b')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.scatter(x,y, color='r', marker='*')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
categories=['A','B','C','D']
values=[10,20,25,30]
plt.bar(categories, values, color='darkcyan')

plt.xlabel("categories")
plt.ylabel("values")
plt.title("bar chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data=np.random.randn(1000)
plt.hist(data, bins=30, color='purple', edgecolor='Black')
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
sizes=[20,30,25,25]
labels=['JS','JAVA','DSA','Angular']
colors=['blue','cyan','pink','purple']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.show()
```
