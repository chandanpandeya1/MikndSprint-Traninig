```python
import numpy as np
import matplotlib.pyplot as plt
sales=np.array([1300,1500,1600,2000,2900,1900,2332,2100,7777,1000])
days=np.arange(1,11) #array of 1 to 10
print(sales)
```

```python
total_sales = np.sum(sales)
average_sales = np.mean(sales)
highest_sales = np.max(sales)
lowest_sales = np.min(sales)

print("Total Sales: ", total_sales)
print("Average Sales: ", average_sales)
print("Highest Sales: ", highest_sales)
print("Lowest Sales: ", lowest_sales)
 
```

```python
plt.figure(figsize=(8,5))
plt.plot(days,sales,marker='o', linestyle='-', color='c', label="Sales")
plt.axhline(average_sales, color='r', linestyle='--',  label=" Average Sales")

plt.xlabel("Days")
plt.ylabel("Sales Revenue ($)")
plt.title("Sales trend over 10 days")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(days,sales, color='c', label="Sales")
plt.axhline(average_sales, color='r', linestyle='--',  label=" Average Sales")

plt.xlabel("Days")
plt.ylabel("Sales Revenue ($)")
plt.title("Sales trend over 10 days")
plt.legend()
plt.grid(axis='y',linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(sales, labels=days,colors=plt.cm.Paired.colors,autopct='%1.1f%%')

plt.title("Sales contribution per day")
plt.show()
```
